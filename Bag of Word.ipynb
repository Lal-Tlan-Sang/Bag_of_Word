```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
#Data Pre-processing
!pip install PorterStemmer
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
```

```python
for i in range(0,len(advices)):
    print(i)
    title = re.sub('[^a-zA-Z]',' ',advices['Advice'][i])
    title = title.lower()
    title = title.split()
    
    title = [ps.stem(word) for word in title if not word in stopwords.words('english')]
    title = ' '.join(title)
    corpus.append(title)
```

```python
corpus
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow = CountVectorizer().fit(corpus)
article = bow.transform(corpus)
count_token = bow.get_feature_names()
count_token
```

```python
df = pd.DataFrame(data =article.toarray(),columns = count_token)
```
